# Importar bibliotecas

````
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

# Graficar ecuaci√≥n lineal en dos dimensiones

````
x = np.array([1, -1])
y = 2*x + 4
plt.plot(x, y)
````

## A√±adir informaci√≥n al gr√°fico

````
plt.plot(x, y, label ='2x+4')
plt.title ('Mi primer funci√≥n en Python')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid(alpha=.8, linestyle='--')

plt.legend()
plt.show()
````

## Grafica de dos ecuaciones

````
plt.plot(x, y, label ='2x+4')
plt.plot(x, -3*x+2, label ='-3x+2')
plt.title ('Mi primer funci√≥n en Python')
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.grid(alpha=.8, linestyle='--')

plt.legend()
plt.show()
````

# Soluci√≥n a un sistema de ecuaciones con dos variables

Supongamos que queremor resolver y luego graficar el siguiente sistema de ecuaciones
```
-3x + y = 4
 2x + y = 1
```




````
#Matriz de coeficientes
M = np.array([[-3, 1], [2, 1]]) 

# vector con las cantidades al lado derecho de las ecuaciones
b =  np.array([4, 1])

solucion = np.linalg.solve(M, b)
solucion
````

## Soluci√≥n gr√°fica 

````
x = np.linspace(-4, 4, 100) 

#despejamos la y
y1 =  3*x + 4
y2 = -2*x + 1 

fig = plt.figure(figsize = (6, 4))

#crear grafico de linea que representa las ecuaciones
plt.plot(x, y1, label ='y = 3x + 4')      # ecuaci√≥n 1
plt.plot(x, y2, label = 'y = -2x+1')  # ecuaci√≥n 2 
plt.plot(-0.6, 2.2, marker='o', label= (solucion[0], solucion[1]))

#t√≠tulo
plt.title ('Grafica de distintas funciones')

#etiquetas de los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#cuadr√≠cula
plt.grid(alpha=.8, linestyle='--')

#limites de los ejes
plt.xlim(-3,3)
plt.ylim(-5,5)


#agregar etiqueta de valores
plt.legend( )

#mostrar la gr√°fica
plt.show()
print('x =',solucion[0],
      'y =', solucion[1])
````

#  Solucion a un sistemas de ecuaciones con tres variables

Supongamos que queremos resolver y luego graficar el siguientes sistemas de ecuaciones:

```
4x -2y + z =11
-2x +4y -2z = -16
x -2y + 4z = 17
```


Eliminando ```z``` de las primeras dos ecuaciones ```x = 1```:

y encontramos la linea de intersecci√≥n de los primeros dos planos (ecuaciones) es igual a ```z = 7 + 2y```


````
#Matriz de coeficientes
M = np.array([[4., -2., 1.], [-2., 4., -2.], [1., -2., 4.]]) 

# vector con las cantidades al lado derecho de las ecuaciones
b =  np.array([11, -16, 17])

solucion = np.linalg.solve(M, b)
solucion
````

## Soluci√≥n gr√°fica 

````
fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection='3d')

#vamos a graficar a z en funci√≥n de x y y, para ello creamos primero los planos x, y

x, y = np.linspace(-8,8,50), np.linspace(-8,8,50)
X, Y = np.meshgrid(x, y)

# despejamos en las tres ecuaciones a Z
Z1 = 11 - 4*X + 2*Y
Z2 = (16 - 2*X + 4*Y) / 2
Z3 = (17 - X + 2*Y) / 4

#superficie de la ecuaci√≥n 1
ax.plot_surface(X, Y, Z1, alpha=.6, color='cyan',) 

 #superficie de la ecuaci√≥n 2
ax.plot_surface(X, Y, Z2, alpha=.6, color='yellow')

#linea de intetersecci√≥n entre la superficies de las primeras dos ecuaciones
ax.plot((1, 1),(-8, 8), (-9,23), lw=3,  c='blue')

#superficie de la ecuaci√≥n 3
ax.plot_surface(X, Y, Z3, alpha=.3, color='red')

#soluci√≥n
ax.plot(xs=[1],ys=[-2], zs=[3], c='black',  marker='o')  

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
````
